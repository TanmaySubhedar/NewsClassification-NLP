```python
import pandas as pd
import nltk
```

```python
nltk.download('punkt')
```

```python
fake = pd.read_csv('fake.csv')
genuine = pd.read_csv('true.csv')
```

```python
fake
```

```python
genuine
```

```python
fake['genuinity'] = 0
genuine['genuinity'] = 1
```

```python
data = pd.concat([fake,genuine],axis = 0)
data
```

```python
data = data.reset_index(drop=True)
data
```

```python
data= data.drop(['title','date','subject'], axis = 1)
data
```

# preprocessing

```python
#tokenization
from nltk.tokenize import word_tokenize
data['text'] = data['text'].apply(word_tokenize)
```

```python
data
```

```python
#stemming
from nltk.stem.snowball import SnowballStemmer
ss = SnowballStemmer('english')
```

```python
def stemmer(text):
    return [ss.stem(word) for word in text]
```

```python
data['text'] = data['text'].apply(stemmer)
```

```python
data
```

```python
#stopword removal

def stopword_rem(text):
    return[word for word in text if len(word)>>2]
data['text'] = data['text'].apply(stopword_rem)
```

```python
data
```

```python
data['text'] = data['text'].apply(' '.join)
```

```python
data
```

```python
#splitting the data set into test and train

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data['text'], data['genuinity'],test_size = 0.30)
```

### vectorisation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf =  TfidfVectorizer(max_df =0.7)

tfidf_train = tfidf.fit_transform(X_train)
tfidf_test = tfidf.transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression as LR
model1 = LR(max_iter =900)
model1.fit(tfidf_train, y_train)
```

```python
pred1 = model1.predict(tfidf_test)
```

```python
from sklearn.metrics import accuracy_score
cr1 = accuracy_score(y_test,pred1)
cr1
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
model2 = PassiveAggressiveClassifier(max_iter =100)
model2.fit(tfidf_train, y_train)
```

```python
pred2 = model2.predict(tfidf_test)
```

```python
cr2 = accuracy_score(y_test, pred2)
cr2
```
